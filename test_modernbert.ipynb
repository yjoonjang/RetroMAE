```python
from transformers import AutoModelForMaskedLM
model_name = "answerdotai/ModernBERT-base"
```

```python
import torch

modernbert_model = AutoModelForMaskedLM.from_pretrained(model_name, attn_implementation="flash_attention_2", dtype=torch.bfloat16)
modernbert_model
```

```python
modernbert_model.model.layers[0].attn.Wqkv
```

```python
modernbert_model.to("cuda")
```

```python
from transformers import AutoTokenizer, AutoModelForMaskedLM

model_id = "answerdotai/ModernBERT-base"
tokenizer = AutoTokenizer.from_pretrained(model_id)

text = "The capital of France is [MASK]."
inputs = tokenizer(text, return_tensors="pt").to("cuda")
outputs = modernbert_model(**inputs, )
outputs
```
